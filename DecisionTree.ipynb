```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
dataset.target
```

```python
X=df.iloc[:,:-1]
y=dataset.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier=DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,15))
tree.plot_tree(treeclassifier)
```

### Post prunning 

```python
treeclassifier=DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,15))
tree.plot_tree(treeclassifier,filled=True)
```

### Prediction 

```python
ypred=treeclassifier.predict(X_test)
```

```python
ypred
```

```python
from sklearn.metrics import classification_report,accuracy_score
cscore=classification_report(ypred,y_test)
ascore=accuracy_score(ypred,y_test)
```

```python
print(cscore)
```

```python
print(ascore)
```

### Pre prunning

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter={
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth' :[1,2,3,4,5,0],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifer=DecisionTreeClassifier()
clf=GridSearchCV(treeclassifer,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred=clf.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```
